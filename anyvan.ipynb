```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('data/gb.csv', usecols=['city', 'lat', 'lng', 'population_proper'])
print(cities.head())
```

```python
# Plotting the cities on a scatter plot
fig = plt.figure(figsize=(10, 10))
plt.scatter(cities['lng'], cities['lat'], s=cities['population_proper'] / 1000, alpha=0.5)
plt.title('UK Cities by Population')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
