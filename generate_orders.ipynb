```python
import pandas as pd
import numpy as np
```

```python
orders = []

for o in range(500):
    f = np.random.randint(0, 257)
    t = np.random.randint(0, 257)
    s = np.random.choice([10, 50, 100, 200])
    
    orders.append((o, f, t, s))

orders_df = pd.DataFrame(orders, columns=['order_id', 'from', 'to', 'size'])
```

```python
orders_df.head()
```

```python
orders_df.to_csv('data/orders.csv', index=False)
```
