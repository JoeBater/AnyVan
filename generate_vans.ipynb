```python
import pandas as pd
import numpy as np
```

```python
vans = []
for v in range(100):
    h = np.random.randint(0, 257)
    s = np.random.choice([100, 200, 400, 800])
    md  = np.round(np.random.randint(100, 501), -1)
    on = np.random.choice([0, 1])
   
    vans.append((v, h, s, md, on))

vans_df = pd.DataFrame(vans, columns=['van_id', 'home_city', 'size', 'max_distance', 'overnights'])
```

```python
vans_df.to_csv('data/vans.csv', index=False)
```
